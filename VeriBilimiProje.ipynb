```python
import pandas as pd
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.tail()
```

```python
df.info()
```

***Eksik Veriler ve Kategorik Değişkenlerle İlgilenmek***

```python
df.isnull().sum()
```

Eksik verileri medyan ile doldurma:

```python
df["Arrival Delay in Minutes"] = df["Arrival Delay in Minutes"].fillna(df["Arrival Delay in Minutes"].median())
```

```python
df.isnull().sum()
```

**Kategorik sütunlar: Gender, Customer Type, Type of Travel, Class, satisfaction (hedef değişken)**

Hedef Değişkeni İkili Hale Getirme:

```python
print(df["satisfaction"].unique())
```

```python
df["satisfaction"] = df["satisfaction"].map({"satisfied": 1, "neutral or dissatisfied": 0})
```

**Diğer Kategorik Değişkenleri Sayıya Dönüştürme**

```python
from sklearn.preprocessing import LabelEncoder
```

Label encoding ile nominal kategorik değişkenleri dönüştürme:

```python
label_cols = ["Gender", "Customer Type", "Type of Travel"]

le = LabelEncoder()
for col in label_cols:
    df[col] = le.fit_transform(df[col])
```

```python
df.dtypes
df.head()
```

Ordinal class:

```python
class_order = {"Eco": 0, "Eco Plus": 1, "Business": 2}
df["Class"] = df["Class"].map(class_order)
```

Özellik ve Hedef Değişkenleri Ayırma:

```python
X = df.drop(["satisfaction", "id", "Unnamed: 0"], axis=1)  # ID ve index anlamlı değil
y = df["satisfaction"]
```

Veriyi Train - Validation - Test Olarak Bölme:

```python
from sklearn.model_selection import train_test_split
```

```python
#test seti (%20)
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

#kalanın %25'i validation seti olacak (0.25 x 0.8 = 0.2 toplamda)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.25, stratify=y_temp, random_state=42)

# %60 - %20 - %20 olarak bölündü
```

***MODEL EĞİTME AŞAMASI:***

1) Random Forest:

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)

y_val_pred = rf.predict(X_val)

print("Accuracy:", accuracy_score(y_val, y_val_pred))
print(confusion_matrix(y_val, y_val_pred))
print(classification_report(y_val, y_val_pred))
```

```python
from sklearn.metrics import accuracy_score, classification_report
import warnings
warnings.filterwarnings('ignore')

def evaluate_model(model, X_train, y_train, X_val, y_val, name=""):
    model.fit(X_train, y_train)
    preds = model.predict(X_val)
    acc = accuracy_score(y_val, preds)
    print(f"\n{name} - Accuracy: {acc:.4f}")
    print(classification_report(y_val, preds))
    return acc
```

2) Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)
evaluate_model(lr, X_train, y_train, X_val, y_val, name="Logistic Regression")
```

3) Decision Tree Classifier:

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
evaluate_model(dt, X_train, y_train, X_val, y_val, name="Decision Tree")
```

4) XGBoost:

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
evaluate_model(xgb, X_train, y_train, X_val, y_val, name="XGBoost")
```

En iyi performand XGBoost modeli.

Test setinde XGBoost’u test etme:

```python
y_test_pred = xgb.predict(X_test)
print(classification_report(y_test, y_test_pred))
```

Feature Importance (Öznitelik Önemi): XGBoost modelinin hangi değişkenlere daha çok "ağırlık" veriyor?

```python
import matplotlib.pyplot as plt
import seaborn as sns

importances = xgb.feature_importances_
features = X.columns

importance_df = pd.DataFrame({"Feature": features, "Importance": importances})
importance_df = importance_df.sort_values(by="Importance", ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x="Importance", y="Feature", data=importance_df)
plt.title("XGBoost Feature Importances")
plt.tight_layout()
plt.show()
```
